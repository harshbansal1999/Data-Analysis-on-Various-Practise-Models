```python
import cifar10
import matplotlib.pyplot as plt
```

```python
cifar10.data_path = "data/CIFAR-10/"
```

```python
cifar10.maybe_download_and_extract()
```

```python
class_names = cifar10.load_class_names()
class_names
```

```python
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn import ensemble
from sklearn.linear_model import LogisticRegression
```

```python
images_train=images_train.reshape(images_train.shape[0],-1)
```

```python
images_train.shape
```

```python
pca=PCA(whiten=True)
pca.fit(images_train)
```

```python
k=0
total=0
currentsum=0
while total<0.95:
    total+=pca.explained_variance_ratio_[k]
    k=k+1
k
```

```python
pca=PCA(n_components=k)
images_train_pca=pca.fit_transform(images_train)
images_train_pca.shape
```

```python
images_test=images_test.reshape(images_test.shape[0],-1)
```

```python
images_test_pca=pca.transform(images_test)
images_test
```

```python
rf=ensemble.RandomForestClassifier()
rf.fit(images_train,cls_train)
y_pred=rf.predict(images_test)
```
