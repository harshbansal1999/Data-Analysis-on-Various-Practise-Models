```python
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv("titanic_train.csv",delimiter=",")
x_train=data.iloc[:,:-1]
y_train=data.iloc[:,-1]

def a(b):
    
    if(b=="S"):
        return 0
    elif(b=="C"):
        return 1
    else:
        return 2

    


x_train.Cabin=x_train.Cabin.fillna(0)


def f(s):
    if(s=="female"):
        return 1
    else:
        return 0

x_train.drop("Name",axis=1,inplace=True)
x_train.drop("Cabin",axis=1,inplace=True)
x_train.drop("Ticket",axis=1,inplace=True)
x_train["Embarked"]=x_train.Embarked.apply(a)
x_train["Gender"]=x_train.Sex.apply(f)
x_train.drop("Sex",axis=1,inplace=True)
x_train.Age=x_train.Age.fillna(0)

x_train
```

```python
x_test=pd.read_csv("titanic_test.csv",delimiter=",")

x_test.drop("Name",axis=1,inplace=True)
x_test.drop("Cabin",axis=1,inplace=True)
x_test.drop("Ticket",axis=1,inplace=True)
def f(s):
    if(s=="female"):
        return 1
    else:
        return 0
x_test["Gender"]=x_test.Sex.apply(f)
x_test.drop("Sex",axis=1,inplace=True)
def a(b):
    if(b=="S"):
        return 0
    elif(b=="C"):
        return 1
    else:
        return 2
x_test["Embarked"]=x_test.Embarked.apply(a)
x_test.Age=x_test.Age.fillna(0)
x_test
```

```python
clf = LogisticRegression()
clf.fit(x_train, y_train)
```

```python
clf.score(x_train,y_train)
```

```python
def a(b):
    if(b==0):
        b="airplane"
    elif(b==1):
        b="automobile"
    elif(b==2):
        b="bird"
    elif(b==3):
        b="cat"
    elif(b==4):
        b="deer"
    elif(b==5):
        b="dog"
    elif(b==6):
        b="frog"
    elif(b==7):
        b="horse"
    elif(b==8):
        b="ship"
    elif(b==9):
        b="truck"
```

```python
y_predict=clf.predict(x_test)
```
